```python
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister, transpile

q = QuantumRegister(4, "q")
c = ClassicalRegister(4, "c")

qc = QuantumCircuit(q, c)
qc.swap(q[0],q[1])
qc.h(q[1])
qc.x(q[1])
qc.ch(q[1],q[0])
qc.x(q[1])
qc.swap(q[2],q[3])
qc.h(q[3])
qc.x(q[3])
qc.ch(q[3],q[2])
qc.x(q[3])
qc.measure(q,c)

display(qc.draw(output="mpl"))
```
